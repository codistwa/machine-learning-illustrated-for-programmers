```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Generation of random classification data
np.random.seed(0)
n = 50
x1 = np.random.uniform(0, 1, n)
y1 = np.random.uniform(0, 1, n)
x2 = np.random.uniform(0, 1, n)
y2 = np.random.uniform(1, 2, n)

# Separation of the data into two groups according to the classification
x_0 = x1[y1 < x1]
y_0 = y1[y1 < x1]
x_1 = x2[y2 > x2]
y_1 = y2[y2 > x2]

# Separating line
x_sep = np.linspace(0, 1, 100)
y_sep = x_sep

# Display of the classification points and the separation line
plt.scatter(x_0, y_0, color='red')
plt.scatter(x_1, y_1, color='blue')
plt.plot(x_sep, y_sep, color='black')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Random classification points separated by a diagonal line')

plt.show()
```
